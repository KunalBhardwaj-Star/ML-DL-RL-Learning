```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
# kNN Implementation from scratch - for binary classification

class KNNClassifier:
    def __init__(self , k = 3):
        self.k = k

    def _euclidean_dist(self , x1 , x2):
        return np.sqrt(np.sum((x1 - x2) ** 2))
    
    def fit(self , X , y):
        self.X_train = X
        self.y_train = y

    def _predict_one(self , x):
        distances = [self._euclidean_dist(x , x_train) for x_train in self.X_train]
        
        knn_indices = np.argsort(distances)[:self.k] # this fnx give sorted index values
        
        knn_classes = [self.y_train[i] for i in knn_indices]

        majority_class = np.argmax(np.bincount(knn_classes))
        return majority_class

    def predict(self , X):
        y_pred = [self._predict_one(x) for x in X]
        return np.array(y_pred)
```

```python
#Data
X_train = np.array([[1] , [2] , [3] , [4] , [5]])

y_train = np.array([2 , 4, 6, 8, 10])

X_test = np.array([
    [1 , 5],
    [3 , 5],
    [6 , 5]
])

model = KNNClassifier()
model.fit(X_train , y_train)

y_pred = model.predict(X_test)
print(y_pred)
```

```python
#Visualize
X_train_df = pd.DataFrame(X_train , columns = ["x1"])
X_train_df["label"] = y_train

sns.scatterplot(
    data = X_train_df,
    x = "x1",
    hue = "label"
)

X_test_df = pd.DataFrame(X_test , columns = ["x1" , "x2"])

sns.scatterplot(
    data = X_test_df,
    x = "x1",
    y = "x2",
    color = "black",
    marker = "X",
    label = "Test Points"
)
```
