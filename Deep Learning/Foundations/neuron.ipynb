```python
import torch
import torch.nn as nn
```

```python
#input
torch.manual_seed(42) # same as random_features
inputs = torch.tensor([1.0 , 2.0 , 3.0])
```

```python
neuron = nn.Linear(in_features = 3 , out_features = 1)
```

```python
output = neuron(inputs)
```

```python
output
```

```python
# To check weight
neuron.weight
```

```python
# To check bias
neuron.bias
```

```python
# verification by formula
# output --> Summation(WiXi) + bias
# WiXi means Weight and features dot product dot product in ML --> @
neuron.weight @ inputs + neuron.bias
```
