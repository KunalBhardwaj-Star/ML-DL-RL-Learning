```python
class PersonalAssistant:
    def __init__(self):
        print("Hi , I am your AI Assistant. How can I help you ?")

    def ans_query(self):
        question = input("Ask me any thing : ")
        response = client.models.generate_content(
            model="gemini-3-flash-preview",
            contents=question,
            config=types.GenerateContentConfig(
                system_instruction="Act as a personal assistant",
                temperature=0.7,      # Moved inside config
                max_output_tokens=512 # Correct spelling (tokens, plural)
            )
        )
        return response.text.strip()

    def summarize_email(self):
        email = input("Paste your email here : ")
        prompt = f"Summarize this email in 20-30 lines: {email}"
        
        response = client.models.generate_content(
            model="gemini-3-flash-preview", # Use the current stable flash model
            contents=prompt,
            config=types.GenerateContentConfig(
                system_instruction="Act as expert email assistant",
                temperature=0.3,
                max_output_tokens=512
            )
        )
    
        # It's safer to use this to ensure you get the text from the first candidate
        return response.candidates[0].content.parts[0].text.strip()
```

```python
import requests # Simple library for HTTP requests

# 1. Define your detailed prompt
prompt = "Generate an image of a highly detailed futuristic spaceship, flying over a vibrant nebula, realistic, 8k resolution"

# 2. Use Pollinations.ai (Free Image API)
# It generates an image URL that you can simply download
image_url = f"https://image.pollinations.ai/prompt/{prompt.replace(' ', '%20')}"

print(f"--- Generating image at: ---\n{image_url}")

# 3. Download and save the image to a file
response = requests.get(image_url)

if response.status_code == 200:
    with open("spaceship.png", "wb") as f:
        f.write(response.content)
    print("Success! Image saved as 'spaceship.png'")
else:
    print("Error generating the image.")
```

```python
from google import genai
from google.genai import types
from PIL import Image
import io

client = genai.Client(api_key="AIzaSyA3DFAOsGi_WFqytkpssyRWtbfJQBj7jL0")

# 1. Load your local image
raw_image = Image.open("spaceship.png")

# 2. Request the edit
# Use the multimodal model that supports image output
response = client.models.generate_content(
    model="gemini-3.1-flash-image-preview",
    contents=[
        raw_image, 
        "Add a futuristic cyberpunk neon helmet to the person in this photo."
    ],
    config=types.GenerateContentConfig(
        response_modalities=["IMAGE"] # Crucial: Tell it to return an image
    )
)

# 3. Save the result
for part in response.candidates[0].content.parts:
    if part.inline_data:
        edited_image = Image.open(io.BytesIO(part.inline_data.data))
        edited_image.save("edited_photo.png")
        print("Image successfully edited and saved!")
```
